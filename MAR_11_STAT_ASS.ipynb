## Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.
## ANSWER:-

The main difference between a t-test and a z-test lies in the assumptions made about the population variance and the sample size. Here is a breakdown of their differences:

1. Assumptions:

   - T-test: Assumes that the population variance is unknown or not provided. It estimates the population variance using the sample variance.
   
   - Z-test: Assumes that the population variance is known or provided.

2. Sample Size:

   - T-test: Typically used for smaller sample sizes, especially when the sample size is below 30.
   
   - Z-test: Generally used for larger sample sizes, where the Central Limit Theorem applies and the sample size is typically above 30.

Example Scenarios:

1. T-Test: Let's say a researcher wants to compare the mean test scores of two different classes. They randomly select 20 students from each class and collect their test scores. Since the population variance is unknown, the researcher would use a t-test to compare the means of the two groups.

2. Z-Test: Consider a scenario where a researcher wants to assess whether a new drug has a significantly different effect on blood pressure compared to a placebo. They conduct a clinical trial with a large sample size of 200 participants, divided equally between the drug and placebo groups. Since the population variance is known (e.g., from previous studies or established data), the researcher would use a z-test to compare the mean blood pressure between the two groups.

In summary, when the population variance is known and the sample size is relatively large, a z-test is appropriate. On the other hand, when the population variance is unknown or the sample size is small, a t-test is more suitable.

## Q2: Differentiate between one-tailed and two-tailed tests.
## ANSWER:-

One-Tailed Test:

A one-tailed test, also known as a one-sided test, is a hypothesis test where the alternative hypothesis is directional. In other words, it tests for the possibility of a relationship or difference in a specific direction. The critical region or rejection region is located entirely in one tail of the distribution.

For example, consider a study investigating whether a new treatment increases the average response time. The one-tailed hypothesis would state that the new treatment reduces response time, and the alternative hypothesis would be directional, such as "The average response time is significantly lower with the new treatment compared to the control group."

The statistical significance is determined by comparing the test statistic to the critical value associated with the chosen significance level in the corresponding tail of the distribution. The hypothesis is rejected only if the test statistic falls into the critical region of that specific tail.

Two-Tailed Test:

A two-tailed test, also known as a two-sided test, is a hypothesis test where the alternative hypothesis is non-directional. It tests for the possibility of a relationship or difference in any direction. The critical region or rejection region is divided between the two tails of the distribution.

Continuing with the previous example, a two-tailed hypothesis would state that the new treatment has any effect on response time, without specifying a particular direction. The alternative hypothesis for a two-tailed test would be something like "The average response time is significantly different between the groups."

In a two-tailed test, the statistical significance is determined by comparing the absolute value of the test statistic to the critical value associated with the chosen significance level. The hypothesis is rejected if the absolute value of the test statistic exceeds the critical value in either tail of the distribution.

In summary, the main difference between one-tailed and two-tailed tests is in the directionality of the alternative hypothesis and the placement of the critical region. A one-tailed test focuses on a specific direction, while a two-tailed test allows for differences in any direction.

## Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
## ANSWER:-

In hypothesis testing, Type 1 and Type 2 errors refer to the potential mistakes or errors that can occur when making decisions about hypotheses. Here's an explanation of each type of error along with an example scenario:

Type 1 Error:

A Type 1 error, also known as a false positive, occurs when the null hypothesis is incorrectly rejected, even though it is true. In other words, it is the error of concluding there is a significant effect or relationship when there is no actual evidence supporting it.

Example Scenario: Let's say a pharmaceutical company is testing a new drug's effectiveness for reducing blood pressure. The null hypothesis states that the drug has no effect on blood pressure. However, due to random chance or statistical fluctuations, the study finds a significant difference in blood pressure between the drug and placebo groups. If the researchers reject the null hypothesis and conclude that the drug is effective, when, in fact, it has no actual effect on blood pressure, it would be a Type 1 error.

Type 2 Error:

A Type 2 error, also known as a false negative, occurs when the null hypothesis is failed to be rejected, even though it is false. In other words, it is the error of failing to identify a significant effect or relationship when it exists.

Example Scenario: Continuing from the previous example, let's say the new drug actually has a significant effect on blood pressure, but the study fails to detect it due to a small sample size or other limitations. In this case, if the researchers fail to reject the null hypothesis and conclude that the drug is ineffective when it does have an actual effect on blood pressure, it would be a Type 2 error.

It's important to note that Type 1 and Type 2 errors are inversely related. Reducing the risk of one type of error increases the risk of the other. The significance level (alpha) and the power of the test can be adjusted to control and balance the probability of Type 1 and Type 2 errors based on the specific context and requirements of the study.

## Q4: Explain Bayes's theorem with an example.
## ANSWER:-

Bayes' theorem is a fundamental concept in probability theory and statistics that provides a way to update the probability of an event or hypothesis based on new evidence. It allows us to incorporate prior beliefs or knowledge and revise them in light of observed data. The theorem is stated as:

P(A|B) = (P(B|A) * P(A)) / P(B)

Where:

- P(A|B) represents the probability of event A given event B has occurred.

- P(B|A) represents the probability of event B given event A has occurred.

- P(A) and P(B) represent the probabilities of events A and B independently of each other.

Here's an example to illustrate Bayes' theorem:

Suppose there is a rare disease that affects 1 in every 10,000 individuals in a population. You go to a clinic to get tested for this disease. The test has a 99% accuracy rate, meaning it correctly identifies individuals with the disease 99% of the time and correctly identifies individuals without the disease 99% of the time.

Now, let's apply Bayes' theorem to calculate the probability of having the disease given that the test result is positive:

- P(A) represents the prior probability of having the disease, which is 1 in 10,000 or 0.0001.

- P(B|A) represents the probability of testing positive given that you have the disease, which is the accuracy rate of the test, 99%, or 0.99.

- P(B) represents the probability of testing positive, which can be calculated using the law of total probability.

The law of total probability states that P(B) is the sum of the probabilities of testing positive given having the disease (P(B|A) * P(A)) and testing positive given not having the disease (P(B|¬A) * P(¬A)). Since the test has a 1% false positive rate (correctly identifies individuals without the disease as positive 1% of the time), we can calculate P(B) as follows:

P(B) = (P(B|A) * P(A)) + (P(B|¬A) * P(¬A))

      = (0.99 * 0.0001) + (0.01 * 0.9999)
      
      ≈ 0.010098

Finally, applying Bayes' theorem:

P(A|B) = (P(B|A) * P(A)) / P(B)

         = (0.99 * 0.0001) / 0.010098
         
         ≈ 0.0098 or 0.98%

So, even if the test result is positive, the probability of actually having the disease is only around 0.98%. This example demonstrates how Bayes' theorem allows us to update our prior beliefs (prevalence of the disease) with new evidence (test results) to arrive at a more accurate probability.

## Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
## ANSWER:-

A confidence interval is a range of values that provides an estimate of the unknown population parameter, along with an associated level of confidence. It is used in statistics to quantify the uncertainty or variability of an estimate based on sample data.

To calculate a confidence interval, the following steps are typically followed:

1. Choose a confidence level: The confidence level represents the desired level of certainty or reliability for the interval. Commonly used levels are 95%, 90%, or 99%.

2. Collect and analyze the sample data: Obtain a representative sample from the population of interest and calculate the sample statistic (e.g., mean or proportion).

3. Determine the appropriate critical value: The critical value is based on the chosen confidence level and the distribution associated with the sample statistic. For example, the critical value for a normal distribution is found using the Z-table, while the critical value for a t-distribution depends on the sample size and degrees of freedom.

4. Calculate the margin of error: The margin of error is a measure of the uncertainty and is calculated by multiplying the critical value by the standard deviation (or standard error) of the sample statistic.

5. Construct the confidence interval: Using the sample statistic, the margin of error, and the chosen confidence level, construct the interval by adding and subtracting the margin of error from the sample statistic.

Now, let's consider an example:

Suppose you want to estimate the average height of a certain species of plants. You collect a random sample of 50 plants and measure their heights. The sample mean height is 40 cm, and the sample standard deviation is 5 cm.

To calculate a 95% confidence interval for the population mean height:

1. Choose a confidence level: We select a confidence level of 95%, which corresponds to a significance level (alpha) of 0.05.

2. Collect and analyze the sample data: We collected a sample of 50 plants, measured their heights, and calculated the sample mean height (40 cm).

3. Determine the appropriate critical value: Since the sample size is large (n = 50), we can use the standard normal distribution. For a 95% confidence level, the critical value is approximately 1.96.

4. Calculate the margin of error: The margin of error is determined by multiplying the critical value (1.96) by the standard deviation of the sample mean. In this case, we estimate the population standard deviation using the sample standard deviation (5 cm) divided by the square root of the sample size (√50). The margin of error is approximately 1.38 cm.

5. Construct the confidence interval: We construct the interval by adding and subtracting the margin of error from the sample mean: 40 cm ± 1.38 cm. Thus, the 95% confidence interval for the population mean height is approximately (38.62 cm, 41.38 cm).

Interpretation: We can say with 95% confidence that the true average height of the plants in the population falls within the range of 38.62 cm to 41.38 cm.

The confidence interval provides a range of plausible values for the population parameter based on the sample data and the chosen level of confidence. It helps quantify the uncertainty associated with the estimate and is widely used in statistical inference.

P(A|B) = (P(B|A) * P(A)) / P(B)

P(B) = (P(B|A) * P(A)) + (P(B|¬A) * P(¬A))

P(¬A) represents the complement of having the disease and can be calculated as 1 - P(A).

P(B) = (0.99 * 0.0001) + (0.02 * (1 - 0.0001))

Now we can calculate P(A|B) using Bayes' Theorem:

P(A|B) = (P(B|A) * P(A)) / P(B)

P(A|B) = (0.99 * 0.0001) / P(B)

So, given that a person tests positive for the disease, the probability that they actually have the disease is approximately 0.492%.

This example illustrates how Bayes' Theorem allows us to update the probability of an event (having the disease) based on new evidence (testing positive) and prior knowledge of the event's probability.

## Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.
## ANSWER:-

To calculate the z-score for a data point, we can use the formula:

z = (x - μ) / σ

Where:

- x is the value of the data point,

- μ is the population mean, and

- σ is the population standard deviation.

Let's calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5:

z = (75 - 70) / 5

  = 5 / 5
  
  = 1

The z-score for this data point is 1.

Interpretation:

A z-score of 1 indicates that the data point of 75 is 1 standard deviation above the population mean. In other words, it is located one standard deviation away from the mean in the positive direction. This suggests that the data point is relatively higher or larger compared to the average value in the population.

Given the sample size of 50 participants, we can use a t-test to compare the sample mean to a hypothesized population mean of zero (no weight loss).

The test statistic for a one-sample t-test is calculated as follows:

t = (sample mean - hypothesized mean) / (sample standard deviation / √n)

In this case:

Sample mean (x̄) = 6 pounds

Hypothesized mean (μ0) = 0 pounds

Sample standard deviation (s) = 2.5 pounds

Sample size (n) = 50

Substituting the values:

t = (6 - 0) / (2.5 / √50)

   = 6 / (2.5 / 7.07)
   
   ≈ 16.97

Next, we need to determine the critical value for the t-test. Since we are conducting the test at a 95% confidence level, the critical value will depend on the degrees of freedom (df = n - 1). With n = 50, the critical value can be obtained from a t-distribution table or calculated using statistical software. For a two-tailed test at 95% confidence level, the critical value would be approximately ±2.009.

Comparing the calculated t-value (16.97) to the critical value (±2.009), we can conclude that the test statistic falls within the critical region. Therefore, we reject the null hypothesis and conclude that the weight loss drug is significantly effective at a 95% confidence level.

This indicates that the observed average weight loss of 6 pounds is significantly different from zero, suggesting that the drug has a significant effect on weight loss, based on the provided sample data.
